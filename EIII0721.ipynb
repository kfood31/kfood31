```python
import pandas as pd
final_review = pd.read_excel('/content/drive/MyDrive/review_data2307191600.xlsx')
final_review
```

```python
!pip install transformers
import torch
from transformers import BertTokenizer, BertForSequenceClassification
# KoBERT 모델 로드
model_name = "snunlp/KR-Medium"  # 또는 다른 KoBERT 모델 이름을 사용합니다.
tokenizer = BertTokenizer.from_pretrained(model_name)
model = BertForSequenceClassification.from_pretrained(model_name)
```

```python
predicted_labels = []

for i in final_review['clean_review']:
    inputs = tokenizer(i, return_tensors="pt")
    outputs = model(**inputs)
    logits = outputs.logits
    predicted_class = torch.argmax(logits, dim=1).item()
    predicted_labels.append('부정' if predicted_class == 0 else '긍정')

# 예측 결과를 text 데이터 프레임에 새로운 컬럼으로 추가
final_review['label'] = predicted_labels

print(final_review[['clean_review','label']])
```
